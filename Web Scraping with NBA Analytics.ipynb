# Web Scraping with NBA Analytics

```python
import pandas as pd
```

```python
# The Building Blocks

# Selet a year variable
year =2001

# Get the url you want to scrap
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
```

```python
# Combining the URL + year strings together

url = url_link.format(year)
```

```python
# Read html file from url variable

df = pd.read_html(url, header = 0)
```

DATA CLEANING

```python
#Assigning the first dataframe to another df

df_2022 = df[0]
```

```python
#Finding all the rows that has "Age" in the Age column

df_2022[df_2022.Age == 'Age']
```

```python
# Dropping all the rows that has "Age" in the Age column with the drop function

df= df_2022.drop(df_2022[df_2022.Age == "Age"].index)
```
